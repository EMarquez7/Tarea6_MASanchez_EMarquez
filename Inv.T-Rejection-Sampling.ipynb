#### <span style='color:lightbrown'> 2. MarcoSC08 </span><br>

$$f(x)=30(x^2-2x^3+x^4)$$

$$t(x)=a \sin(\pi x)$$

$$a=max(f(x))$$

$$c=\int_{0}^{1}a sin(\pi x)dx$$
pulling out

$$c=max(f(x))\int_{0}^{1} sin(\pi x)dx$$
$$c= max(f(x))* [-\frac{cos(\pi y)}{\pi} ]_{0}^{1} = \frac{2}{\pi}max(f (x))$$

$$g(x)=\frac{t(x)}{c}=\frac{max(f(x)) \sin(\pi x)}{\frac{2}{\pi}max(f (x))}=\frac{\pi sin(\pi x)}{2}$$

##### Maximization of $t(x) = a\sin(\pi x)$ &nbsp;  <br><br>


We know $t(x) = a \sin(\pi x)$ <br>

$c = \int_{0}^{1} t(x)$<br><br>
<br><br>
<span style='color:lightblue'> We solve by substitution </span> $\color{lightblue}{u = \pi x}$ ,&nbsp; $\color{lightblue}{du= \pi dx}$ <br><br> 

$c= \int_{0}^{1} a \sin(\pi x) dx$ <br><br>
Substitute $\frac{du}{\pi} = dx$<br><br>
We are left with:<br><br>
$c= a \int_{0}^{1} \sin(u) \frac{du}{\pi}$<br><br>
Get the constants out <br><br>
$c= \frac {a}{\pi} \int_{0}^{1} \sin(u) du$<br><br>

Now just solve the integral by substituting back $u = \pi x$<br><br>
$c= \frac{a}{\pi} [-\cos(u)]_{0}^{1}$ <br><br>
$c= \frac{a}{\pi} [-\cos(\pi x)]_{0 }^{1}$<br><br>
Evaluate the limits:<br><br>
$c= \frac{a}{\pi} [-\cos(\pi) + \cos(0)]$<br><br>
$c= \frac{a}{\pi} [-\cos(\pi) + \ cos(0)]$<br><br>
$c= \frac{a}{\pi} [-(-1) + 1]= \frac{2a}{\pi}$<br><br>
<br>
The maximum value on x is $\frac{\pi}{2}$ for $t(x) = a\sin(\pi x)$

### <br><br><br><br> Inverse Transform Method <br>
3. Use the function found in point 2 and use the inverse transform method seen in class 9, to generate random variables that follow the $g(x)$ distribution. <br>
**Note:** Remember that the inverse transform method works with the cumulative probability distribution and not with its density. <br>
Again similar to the previous point report all calculations using Markdown (Latex).

#### <span style='color:lightbrown'> 3. MarcoSC08 </span><br>

$$G(x)= \int_{0}^{x} g(x)dx = \frac{1}{2} \int_{0}^{x} sin(\pi x)$$

$$G(x)=-\frac{1}{2} [cos(\pi x)]^{x}_{0}= \frac{1}{2} (1-cos(\pi x))$$

$$U= \frac{1}{2} (1-cos(\pi x))$$

Clearing...

$$x=2U=(1-cos(\pi x)$$

$$cos(\pi x)+2U =1 $$

$$cos(\pi x)=1 - 2U  $$

$$\pi x= arcos(1 - 2U)  $$

$$x= \frac{arcos(1 - 2U)}{\pi }  $$

##### <span style='color:lightblue'> 3. EstebanMqz </span><br>
First we must integrate by substitution of g(x):

$G(x) = \int_{0}^{x}g(x)$ <br><br>
Substituting $g(x)$ for $t(x)$ and $c$ for $a$: <br><br>
    $= \int_{0}^{x}\frac{\pi \sin(\pi x)}{2}$ <br><br>
Solving the integral: <br><br>
$G(x) = -\frac {1}{2}[cos(\pi x)]_{0}^{x} = -\frac{1}{2}cos(\pi x) + \frac{1}{2}$ <br><br>

Having solved the Inverse transform method analitically, we solve for x: <br><br>
$ G(x) = -\frac{1}{2}cos(\pi x) + \frac{1}{2}= U$ ∴ <br><br>
$ G(x) = -2U + 1 = cos(\pi x) $<br><br>
Therefore, we can obtain x by solving for cos: <br><br>
$$ x = \frac{cos^{-1}(-2U +1)}{\pi}$$

```python
def inverse_transform_sampling(n, G_Inv):
    """
    Generate n samples of a random variable X
    with density function g and cumulative distribution function G.
    Parameters
    ----------
     n: number of samples
    g: density function g
    G_Inv: inverse cumulative distribution function of G
    Returns
    -------
    g_val: array of n samples of X

    """
    u = np.random.rand(n) # random values ​​over u
    g_val = G_Inv(u) # evaluation of inverse transform on u
    return g_val
    
#pdf(x) and cdf(X)
g = lambda x: (np.pi*np.sin(np.pi*x))/2
G_Inv = lambda u: np.arccos(1 - 2*u) / np.pi

def plot_hist(n, g, title):
    """
    Generate n samples of a random variable X
    with density function g and cumulative distribution function G.
    Parameters
    ----------
     n: number of samples
    g: density function g
    G_Inv: inverse cumulative distribution function of G
    Returns
    -------
    g_val: array of n samples of X
    """
    g_val = inverse_transform_sampling(n, G_Inv)
    x = np.linspace(0, 1, 100)
    g_den = g(x)
    #Histogram aesthetics
    plt.style.use('dark_background')
    plt.rc('grid', linestyle="--", color='gray')
    plt. figure(figsize = (12, 8))
    plt.hist(g_val, bins = 50, density = True, color='r', label='Inv.T_Sampling')
    plt.plot(x, g_den, color="w", linewidth=3)
    plt.title(title, size = 18)
    plt.ylabel('Probability', size= 13)
    plt.xlabel('x', size= 13)
    plt.legend()

    return plt.show()
plot_hist(10000, g, 'Inverse Transform Sampling g(x) → G(x)')
```

```python
#pandas set option to display all dataframe rows
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# # Accept and reject function using list compression
def Accept_rejection(f, t, N, title):
    """ Accept_rejection function creates N random values of R1 and R2 within x domain and the prob. density function f(R1) and the validation function t(R2).
    A logical condition with an operator to be used over R2 and the created values of f(R1), t(R1) is established inside the function *(see condition)*.
    It returns a plot of the accepted and rejected values of f(R1), t(R1), the arg. max (a) of the functions f(x) and t(x) and the rejected values as a tuple (x,y)
    Parameters
    ----------
    R2: Random values over the solution of x over the CDF 
    R1: Random values ​​over g(x)
    f: Objective function
    t: Validation function
    N: Number of points to simulate
    condition: Condition to be met
    title: Title of the graph
    Returns
    -------
    scatter plot of the values ​​of R1 and R2 that meet the condition
    a: arg. max of functions f(x) and t(x).
    """
    # Create a list of random numbers between 0 and 1.
    R1 = np.random.rand(N)
    R2 = np.random.rand(N)
    
    # Evaluate the objective and validation functions at the random points.
    f_x = f(R1)
    t_x = t(R1)
    # R2 Random points satisfying the condition.
    condition = R2*t_x <=f_x
    
    # Plot the points.
    plt.figure(figsize = (20, 12))
    [plt.plot(R1[i],R2[i]*t_x[i],'*y') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i], 'xr') for i in range(len(R1))]
    
    # Accepted points.
    x = pd.DataFrame([R1[i] for i in range(len(R1)) if condition[i]]).sort_values(by=0).values
    # Rejected points.
    x_rejected = pd.DataFrame([R1[i] for i in range(len(R1)) if not condition[i]]).sort_values(by=0).values
    
    # Evaluate objective function.
    f_x = f(x)
    
    # Get the points rejected by the function f(x).
    f_x_rejected = f(x_rejected)
    x_f_x_rejected = list(zip(x_rejected, f_x_rejected))
    
    # x,f(x) rejected values as a dataframe
    x_f_x_rejected = pd.DataFrame(x_f_x_rejected, columns=['x', 'f(x)']).sort_values(by='x').values
    x_f_x_rejected = pd.DataFrame(list(zip(*x_f_x_rejected))[0], columns=['x'])
    f_x_rejected = pd.DataFrame(list(zip(*f_x_rejected))[0], columns=['f(x)'])

    rejected_points = pd.concat([x_f_x_rejected, f_x_rejected], axis=1)
    
    # Maximum of objective function
    a = f(optimize.fmin(lambda x:-f(x), 0, disp=False))
    x_max = optimize.fmin(lambda x:-f(x), 0, disp=False)
    t_max = t(x_max) #Maximize the validation function where f(x) is maximized.

    
    
    # Define the validation function
    
    # Plot the objective function, validation function, and the accepted points
    plt.plot(x,f(x),'m',label='$f(x)$', color='white', linewidth=3 , linestyle='dotted')
    plt.plot(x,t(x),'g',label='$t(x)$', color='orange', linewidth=1)
    plt.plot(x_max, t_max, 'o', color='green', markersize=13) # arg. max
    
    # Plot Style
    plt.style.use('dark_background')
    plt.grid()
    plt.title(title,fontsize=18)
    plt.xlabel('x',fontsize=14)
    plt.ylabel('Probability',fontsize=14)
    plt.legend(loc = "best")
    plt.show()
    
    # Return the results
    return print("The arg. max of f(x) and t(x) is: " + str(x_max[0].round(4)) + " , " + str(t_max[0]) + ")" + str(" rejected values are ") + str((len(rejected_points)/N)) + str(":")), rejected_points 

N=500
x = np.arange(0,1,0.01)
f = lambda x:30*(x**2-2*x**3+x**4)
N = 500
a = f ( optimize . fmin ( lambda x : - f ( x ) , 0 , disp = False ) )
x_max = optimize.fmin(lambda x:-f(x), 0, disp=False)
t = lambda x: a * np.sin(np.pi*x)
t_max = t(x_max)


Rejected_values = Accept_rejection(f , t , N, 
"$arg. max_{(x,y)}$ of $t_1(x) = a\sin(\pi x)$ and $f_1(x) = 30(x^2-2x^3+x^4)$ in $x \in$ [0,1] is: " +
 " (" + str(x_max[0].round(4)) + " , " + str(t_max[0]) + ")")[1]

Rejected_values.head()
```
