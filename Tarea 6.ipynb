$$c=\int_{0}^{1}a sin(\pi x)dx$$

$$c=max(f(x))\int_{0}^{1} sin(\pi x)dx$$
$$c= max(f(x))* [-\frac{cos(\pi y)}{\pi} ]_{0}^{1} = \frac{2}{\pi}max(f(x))$$

$$g(x)=\frac{t(x)}{c}=\frac{max(f(x)) \sin(\pi x)}{\frac{2}{\pi}max(f(x))}=\frac{\pi sin(\pi x)}{2}$$

$$c = \int_{0}^{1} a \sin(\pi x) dx = a \int_{0}^{1} \sin(u) \frac{du}{\pi} = \frac{a}{\pi} \int_{0}^{1} \sin(u) du =$$ 


$$\frac{a}{\pi} [-\cos(u)]_{0}^{1} = \frac{a}{\pi} [-\cos(\pi x)]_{0}^{1} = \frac{a}{\pi} [-\cos(\pi) + \cos(0)] = \frac{a}{\pi} [-\cos(\pi) + \cos(0)] = \frac{a}{\pi} [-(-1) + 1]= \frac{2a}{\pi}$$


$$g(x) = \frac{t(x)}{c}$$

$$\frac{p_{i+1}}{p_i} =\frac{p(n-i)!}{(1-p)(i+1)!(n-i-1)!}$$

>si:
$$P(k+1) = \frac{\lambda}{K+1}P(k)$$
> Entonces:
$$\frac{P(k+1)}{P(k)} = \frac{\lambda}{k+1}$$
> en términos de $P(k)$:
$$\frac{P(k+1)}{P(k)} = \frac{\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}}$$ 
> Mediante el uso de la propiedad $$\frac{k!}{(k+1)!}=\frac{1}{k+1}$$
$$\frac{P(k+1)}{P(k)} = \frac{\lambda^{k+1}k!}{\lambda^{k}(k+1)!} = \frac{\lambda}{k+1}$$
Para obtener finalmente que: $$P(k+1) = \frac{\lambda}{k+1}P(k)$$ 

```python
###Función de Poisson de distribución acum.
def dist_acum_poisson(L:"Parámetro de media y varianza",n:"n-esimo término"): 
    "Función acum. de dist. de Poisson"
    vector = np.zeros(n) #Vector de 0s para su llenado
    vector[0] = (np.exp(-L)) #condición inicial
    def llenar_vector(i):
        nonlocal vector #Para el llenao de vector no interno a la función
        vector[i+1] = (L*vector[i])/(i+1) #Llenado iterativo de vector 
    [llenar_vector(i) for i in range(n-1)] #Compresión de listas
    return vector.cumsum() #Función acumulada
def Gen_distr_discreta(p_acum: 'P.Acum. de la distribución',
                       N: 'N Números aleatorios'):
    U =np.random.rand(N)
    v = [sum([1 for p in p_acum if p < ui ])  for ui in U]
    return v
### Función para plotear el histograma de la función de distribución
def plot_histogram_discrete(distribucion:'distribución por histograma',
                            label:'label'):
    plt.figure(figsize=[14,8])
    y,x = np.histogram(distribucion,bins = len(set(distribucion))) 
    plt.bar(list(set(distribucion)),y,label=label, color='darkblue')
    plt.legend()
    plt.show()
    
n=1000 #parametros
L= 4 #lambda 
p_acum = dist_acum_poisson(L,n) #Distribución de poisson
distribucion = Gen_distr_discreta(p_acum,n) #Distribución discreta
plot_histogram_discrete(distribucion,"Dist. de Poisson") #Histograma
```
